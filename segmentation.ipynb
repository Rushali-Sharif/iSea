```python
from ultralytics import YOLO

# Load model
model = YOLO("yolov8n-seg.pt")  # nano segmentation model

# Path to dataset YAML
data_yaml_path = "/kaggle/input/iseaseg/iSea_Segmentation/yaml.yaml"

# Train the model
model.train(data=data_yaml_path, epochs=10, imgsz=640, batch=8)
```

```python
from ultralytics import YOLO

# Load the trained segmentation model
model = YOLO("runs/segment/train/weights/best.pt")
```

```python
from ultralytics import YOLO

# # Load the best model (change path if saved elsewhere)
# model = YOLO("runs/segment/train/weights/best.pt")

# Evaluate on test set (uses test set defined in YAML)
metrics = model.val(data="/kaggle/input/iseaseg/iSea_Segmentation/yaml.yaml", split='test')

# Print key metrics
print("Metrics on test set:")
print(metrics)
```

```python
from ultralytics import YOLO
import os

# Load your trained model
model = YOLO('/kaggle/working/runs/segment/train/weights/best.pt')

# Path to test images
test_images_path = '/kaggle/input/iseaseg/iSea_Segmentation/images/test/'

# Predict on entire test folder
results = model.predict(
    source=test_images_path,  # folder of test images
    imgsz=640,
    save=True,  # this will save predictions in 'runs/segment/predict'
    save_txt=True,  # save labels
    save_conf=True  # save confidences (optional)
)

# Done! You can check the output folder: /kaggle/working/runs/segment/predict
```
