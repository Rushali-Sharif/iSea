```python
from ultralytics import YOLO

# Load nano model
model = YOLO('yolov8n.pt')

# Train
model.train(
    data='/kaggle/input/iseaod/iSea/yaml.yaml',
    epochs=10,
    imgsz=640,
    batch=8,
    name='isea_yolov8n'
)
```

```python
from ultralytics import YOLO

# Load your trained model (make sure the path matches your run)
model = YOLO('/kaggle/working/runs/detect/isea_yolov8n4/weights/best.pt')
```

```python
model.val(data='/kaggle/input/iseaod/iSea/yaml.yaml', split='test')
```

```python
results = model.predict(source='/kaggle/input/iseaod/iSea/test/images', imgsz=640, save=True)
```

```python
# Access save directory from the first result object
pred_dir = results[0].save_dir

```

```python
import os
from IPython.display import Image, display

# Display first 5 predicted images
for img_name in os.listdir(pred_dir)[:5]:
    display(Image(filename=os.path.join(pred_dir, img_name)))
```
